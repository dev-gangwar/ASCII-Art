```python
from logging import raiseExceptions
from cv2 import cv2
import numpy as np
def main():
    vc=cv2.VideoCapture(0)
    if vc.isOpened():
        rval,frame=vc.read()
    else:
        rval=False
    while rval:
        rval,frame=vc.read()
        print(toASCII(frame))
        key=cv2.waitKey(50)
        if key==27:
            break

def toASCII(frame,cols=120,rows=35):
    frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    hieght,width=frame.shape
    cell_width=width/cols
    cell_hieght=hieght/rows
    if cols>width or rows>hieght:
        raise ValueError("cols or rows exceeds limit")
    result=""
    for i in range(rows):
        for j in range(cols):
            gray=np.mean(frame[int(i*cell_hieght):min(int((i+1)*cell_hieght),hieght),int(j*cell_width):min(int((j+1)*cell_width),width)])
            result+=grayToChar(gray)
        result+='\n'
    return result

def grayToChar(gray):
    CHAR_LIST=',:#@!$%^&*}{><'
    num_chars=len(CHAR_LIST)
    return CHAR_LIST[min(int(gray*num_chars/255),num_chars-1)]
if __name__=='__main__':
    main()
```
